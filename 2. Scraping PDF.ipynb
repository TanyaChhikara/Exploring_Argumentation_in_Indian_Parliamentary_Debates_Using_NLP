```python
url = 'https://sansad.in/getFile/Synop/17/I/Supp+Main-Synopsis-21-06-2019.pdf?source=loksabhadocs'
```

```python
import fitz
import requests

response = requests.get(url)
extracted_text = ""

if response.status_code == 200:
    pdf_bytes = response.content

    pdf_document = fitz.open(stream=pdf_bytes, filetype="pdf")

    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)

        text_with_layout = page.get_text("text")

        extracted_text += text_with_layout

    pdf_document.close()
```

### **Summarizing Text**

```python
from gensim.summarization import summarize

summary = summarize(extracted_text)
print("Summary:")
print(summary)
```

### **1. Using LSA Summarizer**

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

r = requests.get(url)
f = io.BytesIO(r.content)

reader = PyPDF2.PdfReader(f)
combined_content = ""

for page_num in range(len(reader.pages)):
    page_content = reader.pages[page_num].extract_text()
    combined_content += page_content

parser = PlaintextParser.from_string(combined_content, Tokenizer("english"))
summarizer = LsaSummarizer()
summary = summarizer(parser.document, sentences_count=15)

print("Summary:")
for sentence in summary:
    print(sentence)
```

### **Extracting member name and corresponding stance**

```python
import re
pattern = r"\(?(?:SHRI|SHRIMATI|DR\.|HON\. SPEAKER)\s(?:[A-Z]+\s?)*:"
matches = re.findall(pattern, extracted_text)
names = [match.split(":")[0] for match in matches]

print("Extracted names:")
for name in names:
    print(name.strip())
```

```python
pattern = r"\(?(?:SHRI|SHRIMATI|DR\.|HON\. SPEAKER)\s(?:[A-Z]+\s?)*:\s(.*?)(?=\s\(?(?:SHRI|SHRIMATI|DR\.|HON\. SPEAKER))"
matches = re.findall(pattern, extracted_text, re.DOTALL)
speaker_speech_dict = {}

for match in re.finditer(pattern, extracted_text, re.DOTALL):
    speaker = match.group(0).split(":")[0].strip()
    speech = match.group(1).strip()
    speaker_speech_dict[speaker] = speech

print("Speaker and Speech Dictionary:")
for speaker, speech in speaker_speech_dict.items():
    print(f"{speaker}: {speech}\n")
```

```python
import pandas as pd

df = pd.DataFrame(list(speaker_speech_dict.items()), columns=['Speaker', 'Speech'])

df.to_csv('speaker_speeches.csv')
```

### **Date of Debate**

```python
import re
sample_text = extracted_text
pattern = r"(Monday|Tuesday|Wednesday|Thursday|Friday|Saturday|Sunday),\s\w+\s\d+,\s\d{4}"
match = re.search(pattern, sample_text)
if match:
    date = match.group()
    print("Extracted date:", date)
else:
    print("Date not found in the text.")
```

```python
import re

sample_text = extracted_text

pattern = r"([A-Z\s]+)\s*\n\s*Secretary General"
match = re.search(pattern, sample_text)
if match:
    secretary_general_name = match.group(1).strip()
    print("Secretary General's Name:", secretary_general_name)
else:
    print("Secretary General's name not found in the text.")
```
